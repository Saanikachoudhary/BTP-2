```python
def int_alpha1(alpha, m):
  intercept = -(3/8)*np.log((256/5)*(np.pi**(1/alpha))*((G*m*m_dot/c**3)**(5/3)))
  return intercept
def int_alpha2(alpha, m):
  intercept = -alpha*np.log((256/5)*(np.pi**(1/alpha))*((G*m*m_dot/c**3)**((1/alpha)-1)))
  return intercept
def int_alpha3(alpha, m):
  intercept = -alpha*np.log((256/5)*(np.pi**(1/alpha))*((G*m*m_dot/c**3)**(5/3)))
  return intercept
```
